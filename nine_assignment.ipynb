Q2.Problem Statement

**Create a 3D cube using the transform property of CSS? **

- Ans :
<!DOCTYPE html>
<html lang="en">
<head>
<meta charset="UTF-8" />
<meta name="viewport" content="width=device-width, initial-scale=1" />
<title>3D CSS Cube</title>
<style>
  body {
    margin: 0;
    height: 100vh;
    background: #121212;
    display: flex;
    justify-content: center;
    align-items: center;
    perspective: 800px;
    font-family: Arial, sans-serif;
  }
  .scene {
    width: 200px;
    height: 200px;
    position: relative;
    transform-style: preserve-3d;
    animation: rotate-cube 10s infinite linear;
  }
  .cube-face {
    position: absolute;
    width: 200px;
    height: 200px;
    background: rgba(0, 150, 250, 0.8);
    border: 2px solid #0096fa;
    box-sizing: border-box;
    line-height: 200px;
    font-size: 40px;
    font-weight: bold;
    color: white;
    text-align: center;
    user-select: none;
  }
  .face-front  { transform: translateZ(100px); background: rgba(255, 80, 80, 0.8); }
  .face-back   { transform: rotateY(180deg) translateZ(100px); background: rgba(80, 255, 80, 0.8); }
  .face-right  { transform: rotateY(90deg) translateZ(100px); background: rgba(80, 80, 255, 0.8); }
  .face-left   { transform: rotateY(-90deg) translateZ(100px); background: rgba(255, 255, 80, 0.8); color: #333; }
  .face-top    { transform: rotateX(90deg) translateZ(100px); background: rgba(255, 80, 255, 0.8); }
  .face-bottom { transform: rotateX(-90deg) translateZ(100px); background: rgba(80, 255, 255, 0.8); color: #333; }

@keyframes rotate-cube {
  from {
    transform: rotateX(0deg) rotateY(0deg);
  }
  to {
    transform: rotateX(360deg) rotateY(360deg);
  }
}
</style>
</head>
<body>
  <div class="scene">
    <div class="cube-face face-front">Front</div>
    <div class="cube-face face-back">Back</div>
    <div class="cube-face face-right">Right</div>
    <div class="cube-face face-left">Left</div>
    <div class="cube-face face-top">Top</div>
    <div class="cube-face face-bottom">Bottom</div>
  </div>
</body>
</html>
